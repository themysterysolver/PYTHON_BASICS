```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy
```

## SINGELTON MEMBERSHIP FUNCTION

```python
x = np.linspace(0, 10, 100)
#print(x)

c = 5 #singelton point

singleton = np.zeros_like(x) #array with similar dim.dtype as x
#print(singleton)
singleton[np.argmin(np.abs(x - c))] = 1 #x-c ==> subtracts argmin finds pos of smallest then we assign it 1!
#print(singleton)


## TESTING

def MembershipSingelton(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = singleton[idx]
  plt.plot(value,memebership_val,'bo')
  return singleton[idx]

for val in [2,5,7]:
  print('The mebership value of {} is {}'.format(val,MembershipSingelton(val)))

# Plot
plt.plot(x, singleton, 'r', linewidth=2) #(X,Y,clr,types...)
plt.title("Singleton Fuzzy Set\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")
plt.ylim(-0.1, 1.1) #ylimitor
plt.show()


```
