```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzzy
```

## SINGELTON MEMBERSHIP FUNCTION

```python
x = np.linspace(0, 10, 100)
#print(x)

c = 5 #singelton point

singleton = np.zeros_like(x) #array with similar dim.dtype as x
#print(singleton)
singleton[np.argmin(np.abs(x - c))] = 1 #x-c ==> subtracts argmin finds pos of smallest then we assign it 1!
#print(singleton)


## TESTING

def MembershipSingelton(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = singleton[idx]
  plt.plot(value,memebership_val,'bo')
  return singleton[idx]

for val in [2,5,7]:
  print('The mebership value of {} is {}'.format(val,MembershipSingelton(val)))

# Plot
plt.plot(x, singleton, 'r', linewidth=2) #(X,Y,clr,types...)
plt.title("Singleton Fuzzy Set\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")
plt.ylim(-0.1, 1.1) #ylimitor
plt.show()


```

## TRIANGULAR MEMBERSHIP FUNCTION

```python
#note here if we increase the value the interval the memebership vlaue of b closes towards peak
x = np.linspace(0,10,1000) #line not LINE
a,b,c = 3,5,8
triMf = fuzzy.trimf(x,[a,b,c])

## TESTING
def MembershipTriangular(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = triMf[idx]
  plt.plot(value,memebership_val,'bo')
  return triMf[idx]

for val in [2,3,4,5,6,7,8,9]:
  print('The mebership value of {} is {} but the actual can be {}'.format(val,MembershipTriangular(val),fuzzy.trimf(np.array([val]),[a,b,c])[0]))


plt.plot(x,triMf,'g-',linewidth=2)
plt.title("Traingular Fuzzy Set\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")
plt.ylim(-0.1, 1.1) #ylimitor
plt.show()
```

## TRAPEZOIDAL MEMBERSHIP FUNCTION

```python
x = np.linspace(0,10,100000)
a,b,c,d = 2,3,5,8
trapMf = fuzzy.trapmf(x,[a,b,c,d])
plt.plot(x,trapMf,'g',linewidth=1)
plt.title("Trapezoidal Fuzzy Set\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")

## TESTING
def MembershipTrapezoidal(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = trapMf[idx]
  plt.plot(value,memebership_val,'bo')
  return trapMf[idx]

for val in [2,3,4,5,6,7,8,9]:
  print('The mebership value of {} is {} but the actual can be {}'.format(val,MembershipTrapezoidal(val),fuzzy.trapmf(np.array([val]),[a,b,c,d])[0]))
```

## R-FUNCTION

## GAUSSIAN MEMBERSHIP

```python
x = np.linspace(0, 10, 100000)
c = 5 #mean
sigma = 1.5 #standard deviation/spread of the curve

gaussMf = fuzzy.gaussmf(x, c, sigma)
plt.plot(x,gaussMf,'m-',linewidth=2)
plt.title("Gaussain Mf\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")

## TESTING
def GaussianMf(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = gaussMf[idx]
  plt.plot(value,memebership_val,'bo')
  return gaussMf[idx]

for val in [0,1,2,3,5,9,10]:
  print('The mebership value of {} is {} but the actual can be {}'.format(val,GaussianMf(val),fuzzy.gaussmf(np.array([val]),c,sigma)[0]))
```

## SIGMOID

```python
x = np.linspace(0, 10, 1000)
c = 5    # crossover point
a = 1.5  # positive
sigMf = fuzzy.sigmf(x, c, a)

plt.plot(x,sigMf,'m-',linewidth=2)
plt.title("Sigmoid Mf\nR.Prabhakara Arjun\n2022503003")
plt.xlabel("Universe")
plt.ylabel("Membership Degree")


## TESTING
def SigmoidMf(value):
  idx = np.argmin(np.abs(x-value))
  memebership_val = sigMf[idx]
  plt.plot(value,memebership_val,'bo')
  return sigMf[idx]

for val in [0,1,2,3,5,9,10]:
  print('The mebership value of {} is {} but the actual can be {}'.format(val,SigmoidMf(val),fuzzy.sigmf(np.array([val]),c,a)[0]))
```
